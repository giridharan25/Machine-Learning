## Prediction using Decision Tree Classifier

### In this exercise we are going create a decision tree classifier and we will visualize the decision tree. The dataset used in this exercise is the iris datset from sklearn datasets.

## Done by : Giridharan S

### Importing necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

### Loading the dataset

```python
from sklearn import datasets
iris = datasets.load_iris()
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

The dependent feature is stored in a variable y.

```python
y = iris.target
y
```

### Model Creation

The model was created and trained using the dataset.

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(df,y)
```

### Visualizing the decision tree classifier model

To visualize the classifier tree, we have to import a library called <b>plot_tree</b>.

```python
from sklearn.tree import plot_tree
```

Before visualizing our tree, we have to create a figure then pharse the model into plot_tree function to visualize our classification tree.

```python
plt.figure(figsize=(17,12))
plot_tree(model,filled=True,feature_names = iris.feature_names,class_names = iris.target_names,fontsize=10)
```

The above figure is our decision tree classifier diagram.
