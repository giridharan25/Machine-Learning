```python
import pandas as pd
from sklearn.datasets import load_digits
import warnings
warnings.filterwarnings('ignore')
```

```python
digit=load_digits()
```

```python
dir(digit)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
```

### GridSearchCV for Random Forest

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf=GridSearchCV(RandomForestClassifier(),{
    'n_estimators':[5,10,15,100,150]
},return_train_score=False,cv=3
)
```

```python
clf.fit(digit.data,digit.target)
```

```python
clf.cv_results_
```

```python
df=pd.DataFrame(clf.cv_results_)
df
```

```python
df[['param_n_estimators','mean_test_score','rank_test_score']]
```

### GridSearchCV for SVM

```python
svm=GridSearchCV(SVC(),{
    'C':[0.5,1,5],
    'kernel':['rbf','linear','sigmoid','poly'],
    'gamma':['scale','auto']
},return_train_score=False,cv=3
)
```

```python
svm.fit(digit.data,digit.target)
```

```python
svm_df=pd.DataFrame(svm.cv_results_)
svm_df
```

```python
svm_df[['param_C','param_gamma','param_kernel','mean_test_score','rank_test_score']]
```

```python
top_5=svm_df[svm_df.rank_test_score<=5][['param_C','param_gamma','param_kernel','mean_test_score','rank_test_score']]
top_5
```

```python
top_5.sort_values(by=['rank_test_score'])
```

```python
svm.best_params_
```

```python
svm.best_score_
```

### To the find the best performaning model with parameters

```python
models={
    'svm':{
        'model':SVC(),
        'params':{
            'C':[0.5,1,5],
            'gamma':['auto','scale'],
            'kernel':['rbf','linear','poly']
        }
    },
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[10,50,100,150,200],
        }
    },
    'logistic':{
        'model':LogisticRegression(),
        'params':{
            'solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
            'multi_class':['auto', 'ovr', 'multinomial']
        }       
    },
    'gaussian':{
        'model':GaussianNB(),
        'params':{}
    },
    'Multinomial':{
        'model':MultinomialNB(),
        'params':{}
    },
    'decisionTree':{
        'model':DecisionTreeClassifier(),
        'params':{
            'criterion':['gini', 'entropy'],
            'splitter':['best','random']
        }
    }
}
```

```python
scores=[]
for model_name,mp in models.items():
    cv=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    cv.fit(digit.data,digit.target)
    scores.append({
        'model':model_name,
        'best_params':cv.best_params_,
        'best_score':cv.best_score_
        }
    )
```

```python
scores
```

```python
score=pd.DataFrame(scores)
score
```

### Therefore SVC {'C': 5, 'gamma': 'scale', 'kernel': 'rbf'} is the best model with 97.3% score.
