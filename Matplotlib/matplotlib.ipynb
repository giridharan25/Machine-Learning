### Axes labels, legends and grid

```python
matches=[1,2,3,4,5]
dhoni=[80,56,70,100,48]
virat=[60,75,40,36,99]
rohit=[40,36,70,50,80]
```

```python
plt.plot(matches,dhoni)
```

```python
plt.plot(matches,dhoni)
plt.plot(matches,virat)
plt.plot(matches,rohit)
```

```python
plt.plot(matches,dhoni,label='Dhoni')
plt.plot(matches,virat,label='Virat')
plt.plot(matches,rohit,label='Rohit')
plt.xlabel('Matches')                            #label for x axis
plt.ylabel('Scores')                             #label for y axis
plt.title('Performance')                         #title for the graph 
plt.legend()                                     #legend is the box that appears on the top left
plt.grid()                                       #grid is the checked background which will be helpful in locating points
```

### Bar Chart

```python
plt.bar(matches,dhoni)
```

```python
players=['Dhoni','Virat','Rohit','Sachin','Yuvaraj']
runs=[100,80,50,60,55]
plt.bar(players,runs,width=0.5)
plt.xlabel('Players')
plt.ylabel('Scores')
plt.title('Performance')
```

```python
matches=[1,2,3,4,5]
dhoni=[80,56,70,100,48]
virat=[60,75,40,36,99]
plt.bar(matches,dhoni,label='Dhoni')
plt.bar(matches,virat,label='Virat')
plt.xlabel('Matches')
plt.ylabel('Scores')
plt.title('Performance')
plt.legend(shadow=True)
```

```python
import numpy as np
xpos=np.arange(len(matches))                   #creating a dummy index for the x axis
xpos
```

```python
plt.bar(xpos-0.2,dhoni,label='Dhoni',width=0.4)  #adding and subtracting a same value will helps to split the plot
plt.bar(xpos+0.2,virat,label='Virat',width=0.4)
plt.xlabel('Matches')
plt.ylabel('Scores')
plt.title('Performance')
plt.xticks(xpos,matches)                         #replacing the dummy indexes with the original data
plt.legend(shadow=True)
```

```python
plt.barh(xpos-0.2,dhoni,label='Dhoni')
plt.barh(xpos+0.2,virat,label='Virat')
plt.title('Performance')
plt.yticks(xpos,matches)
plt.legend()
```

### Histogrom

```python
blood_sugar_men=[90,100,125,150,80,60,110,95,103,89]
blood_sugar_women=[80,96,110,120,130,94,86,90,120,140]
plt.hist(blood_sugar_men,bins=[80,100,120,150],rwidth=0.95,color='lightblue')
```

```python
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,120,150],rwidth=0.95,color=['lightblue','orange'],label=['men','women'])
plt.legend(shadow=True)
plt.grid()
plt.xlabel('Sugar level')
plt.ylabel('No of people')
plt.title('Sugar level report')
```

### Pie chart

```python
expenses=[10000,8000,5000,3000,2000]
exp_labels=['food','loan','entertainment','others','internet']
plt.pie(expenses,labels=exp_labels,autopct='%0.1f%%',radius=1.5,explode=[0.1,0.1,0,0,0],shadow=True,startangle=45)
plt.show()
```
